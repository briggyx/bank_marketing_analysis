## Bootstrap With Replacement, n = 1,000
### Age

```python
n_bootstrap_samples = 1000
sample_size = 45211

bootstrap_samples = np.empty((n_bootstrap_samples, sample_size))

for i in range(n_bootstrap_samples):
    bootstrap_samples[i] = np.random.choice(df['age'], size=sample_size, replace=True)

print(bootstrap_samples)
```

```python
# with open('../data/bootstrap_samples_n1000.pkl', 'wb') as f:
#     pickle.dump(bootstrap_samples, f)
```

```python
file_path = '../data/bootstrap_samples_n1000.pkl'

with open(file_path, 'rb') as f:
    bootstrap_samples = pickle.load(f)
```

```python
bootstrap_means = np.mean(bootstrap_samples, axis=1)
bootstrap_means_df = pd.DataFrame(bootstrap_means, columns=['bootstrap_mean'])
bootstrap_means_df.head()
```

```python
# with open('../data/bootstrap_means_n1000.pkl', 'wb') as f:
#     pickle.dump(bootstrap_means_df, f)
```

```python
with open('../data/bootstrap_means_n1000.pkl', 'rb') as f:
    bootstrap_means = pickle.load(f)
```

```python
bootstrap_means.describe()
```

```python
mean_age = bootstrap_means['bootstrap_mean'].mean()
std_age = bootstrap_means['bootstrap_mean'].std()
min_age = bootstrap_means['bootstrap_mean'].min()
max_age = bootstrap_means['bootstrap_mean'].max()

plt.hist(bootstrap_means, bins=30, alpha=0.7, color='blue', edgecolor='black')

plt.xlabel('Bootstrap Means of Age')
plt.ylabel('Frequency')
plt.axvline(mean_age, color='r', linestyle='dashed', linewidth=1, label=f'Mean: {mean_age:.2f}')
plt.axvline(min_age, color='g', linestyle='dashed', linewidth=1, label=f'Min: {min_age}')
plt.axvline(max_age, color='y', linestyle='dashed', linewidth=1, label=f'Max: {max_age}')
plt.title('Histogram of Bootstrap Means of Age')

plt.show()
```
