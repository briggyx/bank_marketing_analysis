```python
import pandas as pd
import calendar
```

```python
df = pd.read_excel('../data/bank_original.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
for column in df.columns:
    print(df[column].unique())
```

```python
columns_to_dummify = ['job', 'education', 'default', 'housing', 'loan', 'contact', 'poutcome', 'y', 'marital']
df_dummies = pd.DataFrame()

for col in columns_to_dummify:
    dummies = pd.get_dummies(df[col], prefix=col)
    df_dummies = pd.concat([df_dummies, dummies], axis=1)
```

```python
df_dummies.head()
```

```python
df_dummies.to_excel('../data/bank_dummiesonly.xlsx', index = False)
```

```python
# Create a dictionary mapping lowercase month abbreviations to numbers
month_to_number = {month_abbr.lower(): i for i, month_abbr in enumerate(calendar.month_abbr) if month_abbr}
df['month_number'] = df['month'].str.lower().map(month_to_number)
```

```python
df.head()
```

```python
df.info()
```

```python
df.to_excel('../data/bank_nodummies.xlsx', index=False)
```

```python
df_combined = pd.concat([df, df_dummies], axis=1)
```
